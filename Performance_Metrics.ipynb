## A. Compute performance metrics for the given data '5_a.csv'

```python
import numpy as np
import pandas as pd
from tqdm import tqdm
from google.colab import drive
drive.mount('/content/drive/')
```

```python
df_a = pd.read_csv('/content/drive/My Drive/Applied AI Course/Assignments/5_a.csv')
df_a.head()
```

```python
df_a['y_pred'] = np.where(df_a['proba']<0.5,0,1)
df_a.head()
```

```python
def Get_scores(df,y_col,y_pred_col):
  if len(df[df[y_pred_col]==1]) == 0:
    TP = 0
  else:
    TP = len(df[(df[y_pred_col]==1) & (df[y_col]==1)])
  if len(df[df[y_pred_col]==0]) ==0:
    TN = 0
  else:
    TN = len(df[(df[y_pred_col]==0) & (df[y_col]==0)])
  if len(df[df[y_pred_col]==1]) == 0:
    FP = 0
  else:
    FP = len(df[(df[y_pred_col]==1) & (df[y_col]==0)])
  if len(df[df[y_pred_col]==0]) == 0:
    FN = 0
  else:
    FN = len(df[(df[y_pred_col]==0) & (df[y_col]==1)])

  Precision = TP/(TP+FP)
  Recall = TP/(TP+FN)
  F1_score = 2*(Precision*Recall)/(Precision+Recall)
  Accuracy = (TP+TN)/(TP+TN+FP+FN)

  print(' True Positive: {}\n True Negative: {}\n False Positive: {}\n False Negative: {}\n'.format(TP,TN,FP,FN))

  print(' Precision: ',Precision)
  print(' Recall\t: ',Recall)
  print(' F1_score: ',F1_score)
  print(' Accuracy: ',Accuracy)
```

```python
Get_scores(df_a,'y','y_pred')
```

```python
def Get_Thresholds(df,y_pred_col):
  len_1s = 0
  len_0s = 0
  ones = []
  zeros = []

  df = df.sort_values(by=y_pred_col,ascending=False)
  for i in tqdm(range(0,len(df))):

    if i == 0:
      len_1s = 1
    elif len(ones) < i+1:
      len_1s += 1
    for j in range(i,len(df)-1):
      if df.iloc[i][y_pred_col] == df.iloc[j+1][y_pred_col]:
        len_1s += 1
      else:
        break

    if len_1s > len(df):
      len_1s = len(df)
    elif len_1s < 0:
      len_1s = 0
    len_0s = len(df)-len_1s
    if len_0s > len(df):
      len_0s = len(df)
    elif len_0s < 0:
      len_0s = 0

    ones = np.ones(len_1s)
    zeros = np.zeros(len_0s)
    df.insert(len(df.columns), i, (np.rint(np.concatenate([ones,zeros]))).astype(int))

  return(df)
```

```python
df_a_with_Threshold = Get_Thresholds(df_a,'proba')
```

```python
def Get_AUC(df,other_cols_num,y_col):
  tpr = []
  fpr = []
  for i in tqdm(range(0,len(df.columns)-other_cols_num)):

    TP = len(df[(df[i]==1) & (df[y_col]==1)])
    TN = len(df[(df[i]==0) & (df[y_col]==0)])
    FP = len(df[(df[i]==1) & (df[y_col]==0)])
    FN = len(df[(df[i]==0) & (df[y_col]==1)])

    if (TP+FN) == 0:
      tpr.append(0)
    else:
      tpr.append(TP/(TP+FN))
    if (FP+TN) == 0:
      fpr.append(0)
    else:
      fpr.append(FP/(FP+TN))

  print('\n\nAUC: ',np.trapz(tpr,fpr))
```

```python
Get_AUC(df_a_with_Threshold,3,'y')
```

```python
import numpy as np
import pandas as pd
from tqdm import tqdm
from google.colab import drive
drive.mount('/content/drive/')
```

```python
df_c = pd.read_csv('/content/drive/My Drive/Applied AI Course/Assignments/5_c.csv')
df_c.head()
```

```python
df_c['y_pred'] = np.where(df_c['prob']<0.5,0,1)
df_c.head()
```

```python
def Get_Thresholds(df,y_pred_col):
  len_1s = 0
  len_0s = 0
  ones = []
  zeros = []

  df = df.sort_values(by=y_pred_col,ascending=False)
  for i in tqdm(range(0,len(df))):

    if i == 0:
      len_1s = 1
    elif len(ones) < i+1:
      len_1s += 1
    for j in range(i,len(df)-1):
      if df.iloc[i][y_pred_col] == df.iloc[j+1][y_pred_col]:
        len_1s += 1
      else:
        break

    if len_1s > len(df):
      len_1s = len(df)
    elif len_1s < 0:
      len_1s = 0
    len_0s = len(df)-len_1s
    if len_0s > len(df):
      len_0s = len(df)
    elif len_0s < 0:
      len_0s = 0

    ones = np.ones(len_1s)
    zeros = np.zeros(len_0s)
    df.insert(len(df.columns), i, (np.rint(np.concatenate([ones,zeros]))).astype(int))

  return(df)
```

```python
df_c_with_Threshold = Get_Thresholds(df_c,'prob')
```

```python
def Get_Final_Threshold(df,other_cols_num,y_col,y_prob_col):

  num_of_FP = []
  num_of_FN = []

  for i in tqdm(range(0,len(df.columns)-other_cols_num)):
    num_of_FP.append(len(df[(df[i]==1) & (df[y_col]==0)]))
    num_of_FN.append(len(df[(df[i]==0) & (df[y_col]==1)]))

  A = []
  for i in range(0,len(num_of_FP)):
    weighted_err = (500*num_of_FN[i]) + (100*num_of_FP[i])
    A.append(weighted_err)
  lowest_err_idx = np.argmin(A)

  return(df.iloc[lowest_err_idx][y_prob_col])
```

```python
Final_Threshold = Get_Final_Threshold(df_c_with_Threshold,3,'y','prob')
print('\nFinal_Threshold: ',Final_Threshold)
```

```python
df_c['y_with_threshold'] = np.where(df_c['prob']<Final_Threshold,0,1)
df_c.sort_values(by='prob',ascending=False).head(10)
```

## D.</b></font> Compute performance metrics(for regression) for the given data 5_d.csv

```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive/')
```

```python
df_d = pd.read_csv('/content/drive/My Drive/Applied AI Course/Assignments/5_d.csv')
df_d.head()
```

```python
MSE = 0
mean_pred = 0
SS_res = 0
SS_tot = 0
for i in range(0,len(df_d)):
  MSE += (df_d['y'][i] - df_d['pred'][i])**2
  mean_pred += df_d['y'][i]
SS_res = MSE

mean_pred = mean_pred/len(df_d)
for j in range(0,len(df_d)):
  SS_tot += (df_d['y'][j] - mean_pred)**2
R_2 = 1 - (SS_res/SS_tot)

MSE = MSE/len(df_d)

print('Mean Squared Error: ',MSE)
print('R^2: ',R_2)
```

```python
MAPE = 0
mean_y = df_d['y'].mean()
for i in range(0,len(df_d)):
  if df_d['y'][i] != 0:
    MAPE += abs((df_d['y'][i] - df_d['pred'][i]) / mean_y)
MAPE = (MAPE/len(df_d)) * 100
print('Mean Absolute Percentage Error: ',MAPE)
```
